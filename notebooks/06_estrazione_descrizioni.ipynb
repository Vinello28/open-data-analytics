# Estrazione Descrizioni per Labeling

Questo notebook espande e processa il file `ai_records_completo.csv` per estrarre la descrizione dei progetti e preparare un dataset contenente una colonna `label` vuota, ideale per le successive fasi di etichettatura (labeling).

```python
import pandas as pd
from pathlib import Path

# Definizione dei percorsi
DATA_DIR = Path('../data')
INPUT_FILE = DATA_DIR / 'ai_records_completo.csv'
OUTPUT_FILE = DATA_DIR / 'descrizioni_da_etichettare.csv'

# Caricamento dataset
# Utilizziamo low_memory=False nel caso ci siano campi misti
df = pd.read_csv(INPUT_FILE, low_memory=False)

print(f"Caricate {len(df)} righe originali.")
df.head(3)
```

```python
# Estraiamo solo la colonna DESCRIZIONE_PROGETTO
df_res = df[['DESCRIZIONE_PROGETTO']].copy()

# Aggiungiamo la colonna 'label' vuota
df_res['label'] = ""

# (Opzionale) Rimuoviamo descrizioni nulle e deduplichiamo
df_res = df_res.dropna(subset=['DESCRIZIONE_PROGETTO'])
df_res = df_res.drop_duplicates(subset=['DESCRIZIONE_PROGETTO'])

print(f"Record validi e unici da analizzare: {len(df_res)}")
display(df_res.head())
```

```python
# Esportiamo in un nuovo file CSV
df_res.to_csv(OUTPUT_FILE, index=False)
print(f"CSV salvato con successo in: {OUTPUT_FILE}")
```
